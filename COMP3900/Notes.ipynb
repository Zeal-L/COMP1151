# User Stories
- 用户故事可以帮助敏捷软件开发团队从终端用户的角度对需求进行简化和高水平的描述。
- 一个用户故事通常遵循以下模板： 
  - As a [who] I want to [what] so that [why]
  - 例子： 作为一个网上购物者，我想把一个项目添加到我的购物车中，这样我就可以购买它了

- 检查表
  - 保持简短
  - 保持简单
  - 从用户的角度写故事
  - 明确故事的原因/价值/好处
  - 只描述一个功能
  - 作为一个团队来写故事
  - 使用验收标准来显示最小可行产品 Minimum Viable Product （MVP），也就是说，是一个可以工作的、可用的产品。

- Acceptance Criteria
  - 验收标准，也叫满足条件，提供了最终用户要求的详细范围
  - 帮助开发团队理解用户故事的价值，并对团队何时应考虑完成某项工作设定期望。
  - 验收标准的目标
  - 澄清团队在开始之前应该建立什么
  - 确保每个人对问题有一个共同的理解
  - 帮助团队成员了解故事的完成时间
  - 通过自动测试帮助验证故事
  - 验收标准应包括：
    - 功能的负面情况
    - 功能性和非功能性的用例
    - 性能方面的关注和准则
    - 系统/功能打算做什么
    - 一个用户故事对其他功能的影响
    - 用户体验问题

  - 例子：
    - 作为一个网上银行的客户，我想要一个强大的密码、这样我的信用卡信息才会安全
    - Acceptance Criteria：
    - 密码必须至少有八（8）个字符
    - 密码必须至少包含以下每组中的一个字符：
      - 小写字母
      - 大写字母
      - 数字
      - 特殊字符（！、@、#、$、%、^、&、*）。


# 敏捷软件开发宣言
- 我们一直在实践中探寻更好的软件开发方法，
- 身体力行的同时也帮助他人。由此我们建立了如下价值观：
  - 个体和互动 高于 流程和工具
  - 工作的软件 高于 详尽的文档
  - 客户合作 高于 合同谈判
  - 响应变化 高于 遵循计划
- 也就是说，尽管右项有其价值，我们更重视左项的价值

# Scrum Agile Methodology
- Scrum是一个敏捷的过程，使我们能够专注于在最短的时间内交付最高的商业价值
- 它允许我们快速、反复地检查实际工作的软件（每两周到一个月）。
- 企业设定优先级。我们的团队自我管理，以确定交付最高优先级功能的最佳方式
- 每两周到一个月，任何人都可以看到实际工作中的软件，并决定按原样发布或继续加强，以便进行另一次迭代
- Sprints 冲刺
  - Scrum项目在一系列的 "冲刺 "中取得进展
  - 目标时间是一到两个星期
  - 恒定的持续时间会带来更好的节奏感
  - 产品在冲刺期间进行设计、编码和测试
  - 在冲刺阶段不做任何改变
    - 围绕你能承诺在冲刺期间不做改变的时间来计划冲刺的持续时间
