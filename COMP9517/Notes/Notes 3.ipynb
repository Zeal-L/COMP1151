# Deep Learning
- 深度学习是目前广泛应用的人工神经网络技术的集合。
- 深度学习技术主要依赖于大量数据和深度学习架构
- 针对不同类型数据和应用的一些著名范式：
  - `Convolutional Neural Networks (CNNs)` 卷积神经网络
    - 卷积神经网络（CNN）是一种用于处理图像的DNN。

    - 卷积神经网络将尝试在早期层中学习边缘和线条等低级特征，然后学习物体的部分特征，并在随后的层中学习物体的高级表示。

# CNNs （gpt）
- 在卷积神经网络（CNNs）中，通常包含三种类型的层：卷积层（Convolutional Layer）、激活层（Activation Layer）和池化层（Pooling Layer）。其中，卷积层和池化层是CNNs的核心组成部分，而激活层则用于引入非线性变换，增强模型的表达能力。

- 卷积层（Convolutional Layer）是CNNs的核心组成部分之一，它通过卷积操作提取输入数据的特征。卷积操作可以看作是一种滤波器的操作，它将一个滤波器（也称为卷积核）应用于输入数据的每个位置，计算出对应位置的输出值。卷积层的输出通常被称为特征图（Feature Map），它可以看作是输入数据在卷积核下的响应。

- 激活层（Activation Layer）用于引入非线性变换，增强模型的表达能力。在卷积神经网络中，常用的激活函数包括ReLU、Sigmoid和Tanh等。其中，ReLU（Rectified Linear Unit）是最常用的激活函数之一，它可以将负数部分截断为0，保留正数部分。这样可以使得模型更加稀疏，减少过拟合的风险。

- 池化层（Pooling Layer）用于减小特征图的尺寸，降低模型的复杂度。池化操作通常是在特征图上进行的，它将每个小区域的值进行聚合，得到一个更小的输出值。常用的池化操作包括最大池化（Max Pooling）和平均池化（Average Pooling）等。最大池化操作会选择每个小区域中的最大值作为输出值，而平均池化操作则会计算每个小区域的平均值作为输出值。

# Regularization
- `Weight Decay` 权重衰减
  - 在训练集的损失函数中加入惩罚项，以降低所学模型的复杂度。
  - 权重衰减的常用选择：
    - L1：惩罚的目的是最小化权重的绝对值。
    - L2: 惩罚的目的是最小化权重的平方值
    - 

- `Dropout` 剔除
  - L1和L2通过修改代价函数减少过拟合
  - Dropout 通过在训练过程中随机丢弃神经网络中的神经元来规整网络, 可以使模型的泛化能力更好
  - 

- `Data Preprocessing` 数据预处理
  - 在将图像中的像素值作为深度神经网络的输入进行训练或评估之前，必须对其进行缩放。
  - 三种主要的像素缩放类型：
  - 像素归一化 Normalization：将像素值缩放至0-1之间。
  - 像素居中 Centering：将像素值缩放至平均值为零
  - 像素标准化 Standardization：缩放像素值，使其均值为零，方差为单位。
  - 

- `Batch Normalization` 批量标准化
  - 是一种用于加速深度神经网络训练的技术。它通过对每个小批量数据进行归一化，使得网络中每一层的输入数据分布更加稳定，从而加速网络的收敛速度，提高网络的泛化能力。
  - 在深度神经网络中，每一层的输入数据分布通常是不稳定的，这会导致网络的收敛速度变慢，甚至出现梯度消失或梯度爆炸等问题。Batch Normalization通过对每个小批量数据进行归一化，使得每一层的输入数据分布更加稳定，从而加速网络的收敛速度，提高网络的泛化能力。
  - Batch Normalization的具体实现是在每个小批量数据上计算均值和方差，并对数据进行归一化。
  - 通常在卷积层或全连接层之后使用，并在激活函数之前进行归一化

## Object Detection
- `R-CNN`
  - 非端到端训练
  - 使用预先训练的图像分类网络（在 ImageNet 上）提取生成提案中的特征
  - 对区域进行分类并完善边界框位置（基于建议框）
  - 慢速 每个 RoI（感兴趣区域）的独立前向过程
- `Fast R-CNN`
  - 依靠预先生成的提案作为 R-CNN
  - 在使用提案 bbox 进行裁剪之前，通过 CNN 转发图像
  - 对特征图进行裁剪--RoI 汇集！- 将提案 bbox 应用于图像坐标和特征图
  - RoI align 是 RoI pooling 的 "亚像素 "版本，来自 "Mask R-CNN"。
  - 结束到结束 
- `Faster R-CNN`
  - 在快速 R-CNN 中插入区域建议网络 (RPN)，根据特征预测建议
  - 区域建议网络：预测某个位置是否应该有一个锚 bbox - RPN 的分类损失
  - 在统一框架和网络中生成建议
  - 两阶段方法：生成建议（RPN）和检测
